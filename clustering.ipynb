```python
#ellbow max
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Datensatz laden
data = pd.read_csv('kunden_einkaufszentrum.csv')

# Auswahl der relevanten Features für das Clustering
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]  # Beispiel: Einkommen und Spending Score

# Elbow-Methode: WCSS für verschiedene Cluster-Anzahlen berechnen
wcss = []
for i in range(1, 11):  # Teste Cluster-Anzahlen von 1 bis 10
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Visualisierung der Elbow-Methode mit Matplotlib
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
#kMeans Clustering

from sklearn.cluster import KMeans

# KMeans-Modell erstellen
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)

# Modell trainieren
kmeans.fit(X)

# Cluster-Zuordnung für jeden Datenpunkt berechnen
y_kmeans = kmeans.predict(X)

# Cluster-Zentren berechnen
centers = kmeans.cluster_centers_

# Visualisierung der Cluster
plt.figure(figsize=(8, 6))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.75, label='Cluster Centers')
plt.title('KMeans Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```
