```python
#lineare Regression von Alter und Einkommen
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = pd.read_csv('kunden_einkaufszentrum.csv')

X = data[['Age']]  # Unabhängige Variable (Alter)
y = data['Annual Income (k$)']  # Abhängige Variable (Einkommen)


model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Annual Income (k$)'], color='blue', label='Data Points')
plt.plot(data['Age'], y_pred, color='red', label='Linear Regression Line')
plt.title('Alter vs. Einkommen (Lineare Regression)')
plt.xlabel('Alter (in Jahren)')
plt.ylabel('Einkommen (k$)')
plt.legend()
plt.grid(True)
plt.show()


print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

```python
#Gender vs. Annual Income (Lineare Regression)

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datensatz laden
data = pd.read_csv('kunden_einkaufszentrum.csv')


data['Gender_Numeric'] = data['Gender'].map({'Male': 0, 'Female': 1})

# Daten für die Regression vorbereiten
X = data[['Gender_Numeric']]  # Unabhängige Variable (Gender)
y = data['Spending Score (1-100)']  # Abhängige Variable (Spending Score)

# Lineares Regressionsmodell erstellen und trainieren
model = LinearRegression()
model.fit(X, y)

# Vorhersagen berechnen
y_pred = model.predict(X)

# Visualisierung der linearen Regression
plt.figure(figsize=(8, 6))
plt.scatter(data['Gender_Numeric'], data['Spending Score (1-100)'], color='blue', label='Data Points')
plt.plot(data['Gender_Numeric'], y_pred, color='red', label='Linear Regression Line')
plt.title('Geschlecht vs. Ausgaben (Lineare Regression)')
plt.xlabel('Geschlecht')
plt.ylabel('Ausgaben (1-100)')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.legend()
plt.grid(True)
plt.show()

# Koefizienten der Regression ausgeben
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```
