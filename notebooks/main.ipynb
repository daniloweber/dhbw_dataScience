```python
# Gender vs. Spending Score
import pandas as pd
import matplotlib.pyplot as plt

# Datensatz laden
data = pd.read_csv('../kunden_einkaufszentrum.csv')

# Durchschnittlicher Spending Score basierend auf dem Geschlecht berechnen
avg_spending_by_gender = data.groupby('Gender')['Spending Score (1-100)'].mean()

# Visualisierung: Balkendiagramm
plt.figure(figsize=(8, 6))
avg_spending_by_gender.plot(kind='bar', color=['pink', 'blue'], title='Durchschnittliche Ausgaben nach Geschlecht')
plt.xlabel('Geschlecht')
plt.ylabel('Durchschnittliche Ausgaben (1-100)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
